# How many shoes do people own?

```r
library(tidyverse)
library(extrafont)
source('https://raw.githubusercontent.com/UrbanInstitute/urban_R_theme/master/urban_theme.R')
```

```r
df = read_csv('./files/shoes_results.csv')
```

```r
df %>%
filter(Answer.Shoes_comfortable >= 0) %>%
ggplot(aes(x=Answer.Shoes_comfortable,
                   #fill=Answer.Gender
          )) +
geom_histogram(alpha = .7,
              position = 'dodge', bins = 15) + 
labs(title='How many shoes people own',
     subtitle='Histogram of comfortable pairs of shoes reported by Turkers',
    x='Shoes per person',
    y = 'Count')
```

```r
df %>%
filter(Answer.Shoes_comfortable >= 0) %>%
ggplot(aes(x=Answer.Shoes_comfortable,
                   fill=Answer.Gender
          )) +
geom_histogram(alpha = .7,
              position = 'dodge', bins = 15) + 
labs(title='How many shoes people own',
     subtitle='Histogram of comfortable pairs of shoes reported by Turkers',
    x='Shoes per person',
    y = 'Count')
```
